```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# READ IN CSV
data = pd.read_csv('datasets\clean_data\energy_indicators.csv')
data.head(20)
```

```python
# ASSIGN X AND Y
X = data["Date"].values.reshape(-1, 1)
y = data["Renewable energy extracted by households"].values.reshape(-1, 1)

print("Shape: ", X.shape, y.shape)
```

```python
# PLOT DATA TO SEE IF THERES A LINEAR TREND
plt.scatter(X, y)
plt.xlabel("Year")
plt.ylabel("Renewable energy extracted by households")
```

```python
# USE SKLEARNS TRAIN_TEST_SPLIT TO SPLIT DATA INTO TRAINING AND TESTING
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# CREATE MODEL
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# FIT THE MODEL INTO TRAINING DATA
model.fit(X_train, y_train)
```

```python
# CALCULATE THE MEAN_SQUARED_ERROR AND THE R_SQUARED VALUE

from sklearn.metrics import mean_squared_error, r2_score

# USE MODEL TO MAKE PREDICTIONS
predicted = model.predict(X_test)

# SCORE THE PREDICTIONS WITH MSE AND R2
mse = mean_squared_error(y_test, predicted)
r2 = r2_score(y_test, predicted)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2 ): {r2}")
```

```python
# CALL THE SCORE METHOD AND SHOW THE R2 SCORE
model.score(X_test, y_test)
```
